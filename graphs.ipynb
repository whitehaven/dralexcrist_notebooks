````python
y_false = np.random.normal(120, 10, size=100)
y_true = np.random.normal(100, 10, size=100)
y_indet = np.random.normal(110, 10, size=100)
````

````python
sns.distplot(y_true, hist=False, bins=20, kde=True, label='good')
sns.distplot(y_false, hist=False, bins=20, kde=True, label='bad')
sns.distplot(y_indet, hist=False, bins=20, kde=True, label='can\'t tell')
plt.legend()
````

````python
ttest_ind(y_false, y_true)
````

````python
scipy.stats.ttest_ind(y_false, y_true)
````
